```python
import numpy as np
import pandas as pd
from datetime import datetime
import scipy.stats as st
```

```python
df_libor=pd.read_excel("/home/user/Desktop/env2/VAR/mvp_3/test_sample.xlsx")

df_ARR=df_libor[["Loan ID","Notional","Start Date","Maturity Date","Comp_Freq","Time Duration","ARR ON START DATE","ARR Spread"]]
df_libor.drop(columns=["ARR ON START DATE","ARR Spread"],inplace=True)
df_ARR


#print(type(df['Maturity Date ']))
```

```python
df_libor["Start Date"]= pd.to_datetime(df_libor['Start Date'])
df_libor["Maturity Date"]= pd.to_datetime(df_libor['Maturity Date'])
df_ARR["Start Date"]= pd.to_datetime(df_ARR['Start Date'])
df_ARR["Maturity Date"]= pd.to_datetime(df_ARR['Maturity Date'])
base_rate_calc={"Monthly":"1M","Quarterly":"3M","Six Months":"6M","Annually":"12M"}
time_duration_calc= {"Monthly":1,"Quarterly":3,"Six Months":6,"Annually":12}
effective_col_calc={"Monthly":12,"Quarterly":4,"Six Months":2,"Annually":1}
```

```python
def Interest_Calculation(df,key):
    for i in range(len(df)):
        #print(df.loc[i,"Start Date "].year)
        #df.loc[i,"Time Duration"]=((df.loc[i,"Maturity Date"].year - df.loc[i,"Start Date"].year) * 12 + df.loc[i,"Maturity Date"].month - df.loc[i,"Start Date"].month)/time_duration_calc[df.loc[i,"Comp_Freq"]]
        df.loc[i,"Base Rate Tenor"]=base_rate_calc[df.loc[i,"Comp_Freq"]]
        df.loc[i,"Effective "+str(key)]=(1+(df.loc[i,str(key)+' ON START DATE']+df.loc[i,str(key)+' Spread'])/100)**(1/effective_col_calc[df.loc[i,"Comp_Freq"]])-1
        df.loc[i,str(key)+" Interest"]=round(df.loc[i,"Effective "+str(key)]*df.loc[i,"Notional"],2)
        df.loc[i,str(key)+" Total Interest"]=(df.loc[i,"Notional"]*(1+df.loc[i,"Effective "+str(key)])**df.loc[i,"Time Duration"])-df.loc[i,"Notional"]
    return df
```

```python
df_libor=Interest_Calculation(df_libor,'LIBOR')
df_libor
```

```python
df_ARR=Interest_Calculation(df_ARR,"ARR")
df_ARR
```

```python
df=df_libor
df[["ARR ON START DATE","ARR Spread","Effective ARR","ARR Interest","ARR Total Interest"]]=df_ARR[["ARR ON START DATE","ARR Spread","Effective ARR","ARR Interest","ARR Total Interest"]]
df
```

```python
for i in range(len(df)):
    df.loc[i,"ARR Spread Calc"]= ((((df.loc[i,"Effective LIBOR"]+1)**effective_col_calc[df.loc[i,"Comp_Freq"]])-1)*100)-df.loc[i,"ARR ON START DATE"]
    df.loc[i,"Effective ARR Calc"]= (1+(df.loc[i,'ARR ON START DATE']+df.loc[i,'ARR Spread Calc'])/100)**(1/effective_col_calc[df.loc[i,"Comp_Freq"]])-1
    df.loc[i,"Interest Calc"]= round(df.loc[i,"Effective ARR Calc"]*df.loc[i,"Notional"],2)
    df.loc[i,"Total Interest Calc"]=(df.loc[i,"Notional"]*(1+df.loc[i,"Effective ARR Calc"])**df.loc[i,"Time Duration"])-df.loc[i,"Notional"]
    
df
```

```python
arr=list(df["New ARR"].unique())
ARR_dict={}
for item in arr:
    ARR_dict[item]=[]
for i in range(len(df)):
    ARR_dict[df.loc[i,"New ARR"]].append(df.loc[i,"ARR Spread Calc"])
for i in range(len(df)):
    df.loc[i,"ARR Spread Adj"]=round(np.mean(ARR_dict[df.loc[i,"New ARR"]]),4)
    df.loc[i,"Effective ARR Adj"]= (1+(df.loc[i,'ARR ON START DATE']+df.loc[i,'ARR Spread Adj'])/100)**(1/effective_col_calc[df.loc[i,"Comp_Freq"]])-1
    df.loc[i,"Interest Adj"]= round(df.loc[i,"Effective ARR Adj"]*df.loc[i,"Notional"],2)
    df.loc[i,"Total Interest Adj"]=(df.loc[i,"Notional"]*(1+df.loc[i,"Effective ARR Adj"])**df.loc[i,"Time Duration"])-df.loc[i,"Notional"]
    df.loc[i,"Impact"]= df.loc[i,"Total Interest Adj"] - df.loc[i,"Total Interest Calc"]
    df.loc[i,"EMI ARR"]= (df.loc[i,"Notional"]*df.loc[i,"Effective LIBOR"]*(1+df.loc[i,"Effective LIBOR"])**df.loc[i,"Time Duration"])/(((1+df.loc[i,"Effective LIBOR"])**df.loc[i,"Time Duration"])-1)
df
#df.to_excel("fixed_rate_portfolio.xlsx")
```
